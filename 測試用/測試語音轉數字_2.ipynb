```python
import os
```

```python
import torchaudio
from speechbrain.pretrained import EncoderClassifier
language_id = EncoderClassifier.from_hparams(source="TalTechNLP/voxlingua107-epaca-tdnn", savedir="tmp")
```

```python
if not os.path.exists("mp3"):
    os.makedirs("mp3")
```

### <font color=#FF3030>先把錄音檔放到 "mp3" 資料夾底下後，再去執行之後的程式碼</font>

```python
os.listdir("mp3")
```

```python
for file in os.listdir("mp3"):
    signal = language_id.load_audio(f"mp3/{file}")
    prediction =  language_id.classify_batch(signal)
    print(file, prediction[3], sep="\t")
```

```python
prediction_list = []
emb_list = []

for file in os.listdir("mp3"):
    signal = language_id.load_audio(f"mp3/{file}")
    prediction_list.append(language_id.classify_batch(signal)[3])
    emb_list.append(language_id.encode_batch(signal))
```
