```python
import torch
```

```python
def concat_data(data_type: str) -> None:
    data_1000 = torch.load(f"1000_{data_type}_emb.pt")
    data_4000 = torch.load(f"4000_{data_type}_emb.pt")

    torch.save(data_1000+data_4000, f"5000_{data_type}_emb.pt")
```

```python
for data_type in ["train", "test", "valid"]:
    concat_data(data_type)
```
